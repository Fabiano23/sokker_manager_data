```python
# Para que possamos utilizar as funções criadas em functions.py
import sys  
sys.path.append('/home/fabiano/Documents/Projetos/sokker_manager_data/dados')  

#Manipulação dos dados
from functions import conecta_ao_db
import pandas as pd

#Algoritmos para formar os grupos, validar número de grupos 
#e redução de dimensionalidade para visualização dos grupos
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from kneed import KneeLocator
from sklearn.decomposition import PCA

#Visualização
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
from adjustText import adjust_text

# Tamanho padrão das figuras neste notebook
plt.rcParams["figure.figsize"] = (16,12)
```

```python
# Lê base de dados e pega informações dos jogadores
db= conecta_ao_db('/home/fabiano/Documents/Projetos/sokker_manager_data/dados/dados_sokker.db')
curs= db.cursor()
df = pd.read_sql_query("SELECT * FROM players", db)
```

```python
last_extraction= df['date_extracted'].values[-1]
```

```python
# Cria dataframe com as últimas habilidades dos atletas (antes de serem vendidos ou após o último treinamento)
query= ('ID == @player_id and date_extracted == @last_extraction')
lista= list()

for player_id in df['ID'].unique():
    df_player= df.query(query).head(1)#.sort_values('date_extracted', ascending= False).head(1)
    lista.append(df_player)


df_players= pd.concat([df for df in lista])
df_players= df_players.reset_index()
df_players.drop(columns= ['index'], inplace= True)
```

# Agrupando atletas por posição

```python
# Retira habilidades que não possui extrema relação direta com determinada função no jogo, ou seja, 
#variáveis importantes independente da função

no= ['skillForm', 'skillExperience', 'skillStamina', 'skillTeamwork', 'skillDiscipline', 'skillPace',
 'skillTechnique']

colunas= [coluna for coluna in df_players.columns if coluna.startswith('skill') and coluna not in no]
colunas
```

```python
# Variáveis que serão utilizadas para agruparmos os jogadores
x_model= df_players[colunas]
x_model.head()
```

*Não precisamos padronizar os valores das variáveis como é comum em problemas de cluster. Neste caso, não temos variáveis que possuem intervalos de valores diferentes (ex: salário e peso) e por isso podemos manter a medida original que varia de 0-20 pra cada característica do jogador.*

Vamos iterar sobre o numero de K e utilizar a silhouetta score e a variância para escolher o melhor número de grupos a serem gerados pelo algoritmo.

```python
# Iterando o número de K do Kmeans e visualizando  a inércia para melhor escolher o número de grupos (k) 

kmeans_kwargs = {"init": "random", "n_init": 10, "max_iter": 300, "random_state": 42}

sse = []
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(x_model)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 7), sse)
plt.xticks(range(2, 7))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
# Mesma ideia acima, mas utilizando o conceito de silhouette score
silhouette_coefficients = []

for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(x_model)
    score = silhouette_score(x_model, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 7), silhouette_coefficients)
plt.xticks(range(2, 7))
plt.xlabel("Number of Clusters")
plt.ylabel("silhouette score")
plt.show()
```

```python
# Melhor K usando o método do cotovelo
kl_sse = KneeLocator(range(2, 7), sse, curve="convex", direction="decreasing")
kl_sse.elbow
```

```python
# Melhor k para o coeficiente de silhueta
kl_silhouette = silhouette_coefficients.index(max(silhouette_coefficients)) + 2
kl_silhouette
```

### Vamos pegar aquele que retornou o maior valor de K já que podemos definir um setor do campo em mais setores 
### Exemplo: Zagueiros e Laterais; Meias volantes e meias avançados; Atacantes e Pontas etc.

```python
#best_k= max([kl_sse.elbow, kl_silhouette])
best_k= kl_silhouette
best_k
```

```python
kmeans = KMeans(n_clusters=best_k, **kmeans_kwargs)
kmeans.fit(x_model)
```

### Visualizando agrupamentos gerados

```python
# Inicializa objeto PCA para redução a duas dimensões
pca_2d = PCA(n_components=2)
```

```python
# Reduzindo variáveis das habilidades dos atletas para duas dimensões
PCs_2d = pd.DataFrame(pca_2d.fit_transform(x_model))

# Renomeando colunas
PCs_2d.columns = ["PC1_2d", "PC2_2d"]
```

```python
# Indicando grupo de cada instância no dataframe com as medidas em PCA
PCs_2d['cluster']= kmeans.labels_
```

```python
# Informações que vamos mostrar no gráfico
names= df_players['surname'].values
sizes= [pace * 1.5 for pace in df_players['skillPace'].values]

# Lista que guardará os textos do plot pra que possamos ajustá-los a fim de que não se sobreponham
texts = list()
```

### Gerando plot com os atletas, os grupos aos quais pertencem e sua agilidade (variável essencial no jogo) indicada no tamanho de seus nomes: quanto maior, mais agilidade o atleta possui e vice versa

```python
g = sns.scatterplot(x="PC1_2d", y="PC2_2d", data=PCs_2d, hue= 'cluster', 
                    palette="deep", s=100)

g.set_title('Jogadores do Organized Chaos FC agrupados por suas habilidades')

for i in range(len(df_players)):
    
    player= plt.text(PCs_2d["PC1_2d"].values[i], PCs_2d["PC2_2d"].values[i], names[i], 
 fontdict=dict(alpha=0.5, size= sizes[i]))
    
    texts.append(player)
    
# Ajusta texto para que um nome não fique em cima do outro
adjust_text(texts)  

# Salvando imagem
g.figure.savefig("Imagens/agrupamento_jogadores.png")
```

Vamos utilizar três variáveis pra observamos a cada rodada os melhores atletas nos aspectos forma, resistência e disciplina tática.
