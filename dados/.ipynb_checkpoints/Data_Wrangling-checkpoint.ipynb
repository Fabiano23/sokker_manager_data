```python
import pandas as pd
import numpy as np
from functions import conecta_ao_db
import seaborn as sns
```

```python
db= conecta_ao_db('dados_sokker.db')
curs= db.cursor()
```

```python
df_atletas= pd.read_sql_query("SELECT * FROM players", db).drop(columns= ['countryID', 
                                                                         'teamID', 'youthTeamID'])


df_time= pd.read_sql_query("SELECT * FROM teams", db)[['date_extracted', 
                                                  'trainingFormation', 'trainingType',
                                                 'fanclubMood']]

df_matches= pd.read_csv('playersStats.csv', parse_dates=True)
```

```python
# Pegando ids dos atletas e deixando num dataframe para puxar seus nomes depois
```

```python
ids_atletas= df_atletas[['ID', 'name', 'surname']]
ids_atletas['Nome']= ids_atletas['name'] + ' ' + ids_atletas['surname']
ids_atletas.drop(columns= ['name', 'surname'], inplace= True)
ids_atletas.drop_duplicates(ignore_index= True, inplace= True)

df_atletas.drop(columns= ['name', 'surname'], inplace= True)
```

```python
ids_atletas= dict(zip(ids_atletas['Nome'].values, ids_atletas['ID'].values))
ids_atletas
```

```python
df_matches['Nome']= df_matches['Nome'].apply(lambda x: x.strip().replace('"', ""))
df_matches['Nome'].unique()
```

```python
df_matches['ID']= df_matches['Nome'].map(ids_atletas)
df_matches.dropna(subset= ['ID'], inplace= True)
df_matches.drop(columns= ['Nome'], inplace= True)
```

```python
df_matches['Data']= df_matches['Data'].apply(lambda x: x.strip().replace('"', ""))
df_matches['Posicao']= df_matches['Posicao'].apply(lambda x: x.strip().replace('"', ""))
df_matches['Nota']= pd.to_numeric(df_matches['Nota'])


df_matches.head()
```

```python
df_atletas['date_extracted']= pd.to_datetime(df_atletas['date_extracted'])
df_time['date_extracted']= pd.to_datetime(df_time['date_extracted'])
df_matches['Data']= pd.to_datetime(df_matches['Data'])
```

```python
# Categorizando coluna training type
treinos= {6: 'Des', 8: 'Agil'}
df_time= df_time.replace({'trainingType': treinos}) 
```

```python
df_time.head(3)
```

### Criando coluna que será nosso "id"

```python
df_atletas['week_number']= df_atletas['date_extracted'].dt.strftime('%U')
df_time['week_number']= df_time['date_extracted'].dt.strftime('%U')
df_matches['week_number']= df_matches['Data'].dt.strftime('%U')
```

```python
# Primeira semana estava testando a coleta e peguei o mesmo dado duas vezes. Vamos iniciar da semana seguinte à primeira coleta

df_atletas= df_atletas[df_atletas['week_number'] != '30'].drop(columns= ['date_extracted'])
df_time= df_time[df_time['week_number'] != '30'].drop(columns= ['date_extracted'])
df_matches= df_matches[~df_matches['week_number'].isin(['30', '29', '28'])].drop(columns= ['Data'])
```

```python
df_time.head()
```

```python
df_atletas.head()
```

```python
df_matches.head()
```

```python
df= df_matches.merge(df_time, on= 'week_number', validate= 'many_to_one').dropna(axis= 1)
df.head()
```

```python
df= df.merge(df_atletas, on=['ID', 'week_number'])
```

```python
df.head()
```

```python
skills_cols= [col for col in df.columns if col.startswith('skill')]
del skills_cols[:5]
skills_cols.append('skillStamina')
skills_cols
```

```python
#Somando as habilidades que sobem com treino de cada atleta
df['skills_sum']= df.loc[:, skills_cols].sum(axis= 1)
```

```python
df.corr()['skills_sum'].sort_values(ascending= False).head(10)
```

```python
id_exemplo= df['ID'].values[0]
df_exemplo= df.query('ID==@id_exemplo')
df_exemplo.head()
```

```python
X= df_exemplo.sort_values('week_number', ascending= True)
```
